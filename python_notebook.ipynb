```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import geopandas as gpd
import statsmodels.api as sm
```

```python
sf_locations_df = pd.read_csv("data/sf_locations.csv")
sf_neighborhoods_df = gpd.read_file("data/sf_neighborhoods.geojson")
sf_waymo_estimates_df = pd.read_csv("data/sf_waymo_estimates.csv")
```

```python
sf_waymo_estimates_df.head()
```

```python
# Combine pickup date + pickup time
sf_waymo_estimates_df["pickup_datetime"] = pd.to_datetime(
    sf_waymo_estimates_df["pickup_date"] + " " + sf_waymo_estimates_df["pickup_time"],
    format="%m/%d/%Y %H:%M"
)

# Same for dropoff if needed
sf_waymo_estimates_df["dropoff_datetime"] = pd.to_datetime(
    sf_waymo_estimates_df["dropoff_date"] + " " + sf_waymo_estimates_df["dropoff_time"],
    format="%m/%d/%Y %H:%M"
)
```

```python
# Hour of day (0â€“23)
sf_waymo_estimates_df["pickup_hour"] = sf_waymo_estimates_df["pickup_datetime"].dt.hour  

# Day of week (0=Monday, 6=Sunday)
sf_waymo_estimates_df["pickup_dayofweek"] = sf_waymo_estimates_df["pickup_datetime"].dt.dayofweek  

# Weekday vs weekend
sf_waymo_estimates_df["is_weekend"] = sf_waymo_estimates_df["pickup_dayofweek"] >= 5  

# Time of day buckets
def time_of_day(hour):
    if 5 <= hour < 12:
        return "Morning"
    elif 12 <= hour < 17:
        return "Afternoon"
    elif 17 <= hour < 21:
        return "Evening"
    else:
        return "Night"

sf_waymo_estimates_df["time_of_day"] = sf_waymo_estimates_df["pickup_hour"].apply(time_of_day)
```

```python
# Define haversine function
def haversine(lat1, lon1, lat2, lon2):
    R = 6371  # Earth radius in km (use 3958.8 for miles)
    
    # Convert degrees â†’ radians
    lat1, lon1, lat2, lon2 = map(np.radians, [lat1, lon1, lat2, lon2])
    
    # Differences
    dlat = lat2 - lat1
    dlon = lon2 - lon1
    
    # Haversine formula
    a = np.sin(dlat/2.0)**2 + np.cos(lat1) * np.cos(lat2) * np.sin(dlon/2.0)**2
    c = 2 * np.arcsin(np.sqrt(a))
    
    return R * c  # Distance in km
```

```python
sf_waymo_estimates_df["trip_distance_km"] = sf_waymo_estimates_df.apply(
    lambda row: haversine(
        row["pickup_latitude"], row["pickup_longitude"],
        row["dropoff_latitude"], row["dropoff_longitude"]
    ),
    axis=1
)

# Optional: also store in miles
sf_waymo_estimates_df["trip_distance_miles"] = sf_waymo_estimates_df["trip_distance_km"] * 0.621371
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Scatter plot (distance vs price)
plt.figure(figsize=(8,6))
sns.scatterplot(
    x="trip_distance_miles",
    y="price_usd",
    data=sf_waymo_estimates_df,
    alpha=0.5
)

# Add labels & title
plt.xlabel("Trip Distance (miles)")
plt.ylabel("Price (USD)")
plt.title("Waymo Price vs Trip Distance in San Francisco")
plt.show()
```

```python
X = sf_waymo_estimates_df["trip_distance_miles"]
y = sf_waymo_estimates_df["price_usd"]

X = sm.add_constant(X)  # adds intercept
model = sm.OLS(y, X).fit()
print(model.summary())
```
